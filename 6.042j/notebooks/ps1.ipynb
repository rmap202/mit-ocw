|   | P     | Q     | R     | $$ \lnot (P \lor (Q \land R) ) =  (\lnot P) \land (\lnot Q \lor \lnot R) ) $$ |

|   | P     | Q     | R     | $$ \lnot (P \land (Q \lor R) ) =  (\lnot P) \lor (\lnot Q \lor \lnot R) ) $$ |
|---|-------|-------|--:-:--|-:-:---------------------------------------------------------------------------|
|   | False | False | False |                                                                               |
|   | False | False | True  |                                                                               |
|   | False | True  | False |                                                                               |
|   | False | True  | True  |                                                                               |
|   | True  | False | False |                                                                               |

|   | True  | True  | False |                                                                               |

```python
inputs = [None, None, None]
flags = [False, False] # per statement
    
s_p1 = 'not (A or (B and C))'
s_q2 = '(not A) and ((not B) or (not C))'
s_s1 = 'not (P and (Q or R))'
s_s2 = '(not P) or ((not Q) or (not R))'

for i in range(2**len(inputs)):
    b_str = "{0:b}".format(i)
    while len(b_str) < 3:
        b_str += '0'
    for j in range(0, len(b_str)):
        val = False
        if (b_str[j] == "1"):
            val = True
        inputs[j] = val

    P, Q, R = inputs[0], inputs[1], inputs[2] # cant make generic - it'll do...
    A, B, C = inputs[0], inputs[1], inputs[2]

    p1 = not (A or (B and C))
    q2 = (not A) and ((not B) or (not C))
    s1 = not (P and (Q or R))
    s2 = (not P) or ((not Q) or (not R))

    if ((not (p1 == q2)) and (not flags[0])):
        flags[0] = True

    if ( (not (s1 == s2)) and (not flags[1])):
        flags[1] = True
        
            
if flags[0]:
    print(f"Statement: {s_p1}={s_q1} is false:\n\tA:\t{A}\n\tB:\t{B}\n\tC:\t{C}")
else:
    print(f"Statement: {s_p1}={s_q1} is true.")
    
if flags[1]:
    print(f"Statement: {s_s1}={s_s2} is false:\n\tP:\t{P}\n\tQ:\t{Q}\n\tR:\t{R}")
else:
    print(f"Statement: {s_s1}={s_s2} is true.")
    
```
